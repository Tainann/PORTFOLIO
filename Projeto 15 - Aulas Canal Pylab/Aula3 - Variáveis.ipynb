# Aula 3 - Tipos de Variáveis

Variável: um espaço da memória, como uma 'caixa', que queremos usar ou manipular ao longo do código.

int: um número inteiro sem casas decimais. \\
float: representa números com casas decimais. \\
string: sequência de caracteres.

Ser int ou float afeta drasticamente a forma de armazenamento e o tempo de execução.

```python
v1 = 10
v2 = 5.25
v3 = 'CR7'
v4 = True

v1, v2, v3, v4
```

```python
# Nome das variáveis não pode começar com caracteres especiais ou números
# Deve começar sempre com letra ou underline
# Normalmente se começa com letras minúsculas

type(v1), type(v2), type(v3), type(v4)
```

Transformando para bool

o comportamento básico de bool():

Considerados False:

    Números:
        0 (inteiros) e 0.0 (float) são False.
    Coleções vazias:
        Strings: ""
        Listas: []
        Tuplas: ()
        Dicionários: {}
        Conjuntos: set()
    Objetos especiais:
        None
        False

Considerados True:

    Qualquer número diferente de zero, incluindo negativos (ex.: 1, -1, 0.5).
    Qualquer string não vazia (ex.: "abc", " ").
    Qualquer coleção não vazia (ex.: [1], {"a": 1}).
    Qualquer objeto que não implemente explicitamente um comportamento de falso.

```python
x = []
x = bool(x)
x
```
