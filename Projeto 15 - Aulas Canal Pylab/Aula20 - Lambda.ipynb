# Funções Lambda


As funções lambda em Python são funções anônimas, ou seja, sem nome, que são definidas em uma única linha para realizar tarefas simples.

São úteis quando se precisa de uma função rápida e curta. Dessa forma:

- Uma função lambda é criada usando a palavra-chave lambda, seguida de um ou mais argumentos, e uma expressão:

- Argumentos são os dados de entrada que esta função irá receber e
expressão é o código que será executado quando a função lambda for chamada.

    lambda argumentos: expressão


Características principais:

- Anônimas: Não possuem nome, a menos que você as atribua a uma variável.
- Simples: Só podem conter uma única expressão, que é avaliada e retornada.
- Limitadas: Não suportam declarações ou múltiplas linhas.

Funções temporárias

- As funções lambda são ideais para funções rápidas e descartáveis, em que definir uma função com def seria desnecessário.

ATENÇÃO: Como são anônimas, é mais complicado chamá-las em múltiplos locais do código sem atribuí-las a uma variável.

```python
dobro = lambda x: x * 2
dobro(5)
```

Uso com funções de ordem superior, como map e filter.




```python
numeros = [1, 2, 3, 4]
quadrados = list(map(lambda x: x ** 2, numeros))
quadrados
```

```python
numeros = [10, 15, 21, 33, 40]
pares = list(filter(lambda x: x % 2 == 0, numeros))
pares
```

Lambda pode ser usada com funções como sorted ou sort para definir critérios de ordenação.

```python
pessoas = [("João", 30), ("Maria", 25), ("Ana", 35)]
ordenado_por_idade = sorted(pessoas, key=lambda pessoa: pessoa[1])
ordenado_por_idade
```

As funções lambda também são usadas para criar pequenas expressões matemáticas ou condicionais.

```python
calculo = lambda a, b: a * b if a > b else a + b
calculo(5, 3), calculo(2, 4)
```

Cuidados com as Funções Lambda:

- Legibilidade: Evite funções lambda muito complexas; prefira usar def quando a lógica for difícil de entender.
- Reusabilidade: Lambdas não são reutilizáveis, o que pode dificultar a manutenção do código.
- Depuração: Não possuem nome, tornando difícil rastrear erros em casos complexos.
