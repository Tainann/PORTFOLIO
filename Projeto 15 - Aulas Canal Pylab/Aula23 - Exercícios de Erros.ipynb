#### 2 - O seguinte código está com erro. Identifique qual exceção deve ser tratada para corrigir o problema.

    import math

    try:
        raiz = math.sqrt(-1)
        print(raiz)
    except:
        print('Ocorreu um erro.')


```python
# sqrt não aceita valores negativos.
# ValueError é o tipo de erro.
import math

try:
    raiz = math.sqrt(-1)
    print(raiz)
except:
    print('Ocorreu um erro.')
```

#### 3 - O que será impresso na execução do código abaixo? Explique o motivo.

    try:
        resultado = 10 / 0
    except ZeroDivisionError:
        print("Erro: Divisão por zero.")
    else:
        print("Divisão realizada com sucesso.")
    finally:
        print("Fim do bloco try-except.")

```python
try:
    resultado = 10 / 0
except ZeroDivisionError:
    print("Erro: Divisão por zero.")
else:
    print("Divisão realizada com sucesso.")
finally:
    print("Fim do bloco try-except.")

# Como ocorre um erro, o que está em else não é executado.
# O que está em finally sempre é executado independente de ter acontecido ou não um erro.
```

#### 4 - Quais serão as saídas do código a seguir:
  
    def dividir(numerador, denominador):
      try:
          resultado = numerador / denominador
      except ZeroDivisionError:
          print("Erro: Não é possível dividir por zero.")
      except TypeError as e:
          print(f"Erro: Entrada inválida - {e}. Certifique-se de que os valores são números.")
      else:
          print(f"Resultado da divisão: {resultado}")
      finally:
          print("Tentativa de divisão finalizada.")
       
    # Chamadas da função para testar:
    dividir(10, 2)
    dividir(10, 0)
    dividir("dez", 2)


```python
def dividir(numerador, denominador):
  try:
      resultado = numerador / denominador
  except ZeroDivisionError:
      print("Erro: Não é possível dividir por zero.")
  except TypeError as e:
      print(f"Erro: Entrada inválida - {e}. Certifique-se de que os valores são números.")
  else:
      print(f"Resultado da divisão: {resultado}")
  finally:
      print("Tentativa de divisão finalizada.")

# Chamadas da função para testar:
dividir(10, 2)
dividir(10, 0)
dividir("dez", 2)
```

#### 5 - Para o código a seguir, o que acontecerá se o usuário digitar "uva"?  

    estoque = {
        "maçã": 10,
        "banana": 5,
        "laranja": 8
    }

    try:
        produto = input("Digite o nome do produto: ")
        print(f"Quantidade em estoque: {estoque[produto]}")
    except KeyError:
        print("Erro: O produto digitado não está no estoque.")
    else:
        print("Consulta realizada com sucesso!")
    finally:
        print("Operação finalizada.")


```python
estoque = {
    "maçã": 10,
    "banana": 5,
    "laranja": 8
}

try:
    produto = input("Digite o nome do produto: ")
    print(f"Quantidade em estoque: {estoque[produto]}")
except KeyError:
    print("Erro: O produto digitado não está no estoque.")
else:
    print("Consulta realizada com sucesso!")
finally:
    print("Operação finalizada.")
```
