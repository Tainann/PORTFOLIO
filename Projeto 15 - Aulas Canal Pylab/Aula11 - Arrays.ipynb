# AULA 11 - ARRAYS

Arrays em Python são estruturas de dados que permitem armazenar múltiplos valores do mesmo tipo em uma sequência.

```python
import numpy as np

# Criando um array a partir de uma lista
array_exemplo = np.array([1, 2, 3, 4, 5])
array_exemplo
```

Arrays com NumPy permitem fazer operações matemáticas diretamente em todos os elementos do array, o que facilita cálculos e torna o código mais eficiente.

```python
# Operações vetorizadas em arrays
array_exemplo = np.array([1, 2, 3, 4, 5])
array_exemplo * 2
```

Operações vetorizadas são cálculos realizados em arrays inteiros de uma vez, ao invés de executar um loop que processa cada elemento individualmente. Esse conceito é especialmente comum em bibliotecas como NumPy e é altamente eficiente, pois permite que operações matemáticas sejam aplicadas a cada elemento de um array (ou vetor) simultaneamente.

```python
zeros_array = np.zeros(5)  # Array de cinco zeros
zeros_array
```

```python
array_de_uns = np.ones((3, 3))  # Array 3x3 de uns
array_de_uns
```

```python
random_array = np.random.rand(4)  # Array de 4 números aleatórios entre 0 e 1
random_array
```

```python
sequencia_array = np.arange(0, 10, 2)  # De 0 a 10, de 2 em 2
sequencia_array
```

### Acessando e modificando elementos

```python
arr = np.array([1, 2, 3, 4])
arr[1]
```

```python
arr[1:3]
```

```python
arr[0] = 10  # Define o primeiro elemento como 10
arr
```

### Operações Matemáticas

```python
arr + 1
```

```python
arr * 2
```

```python
arr2 = np.array([4, 5, 6, 8])

arr + arr2  # Soma os elementos correspondentes
```

```python
np.sum(arr)     # Soma de todos os elementos
```

```python
np.mean(arr)   # Média dos elementos
```

```python
np.min(arr)    # Menor valor do array
```

```python
np.max(arr)     # Maior valor do array
```

### Funções Matemáticas

```python
angulos = np.array([0, np.pi/2, np.pi])
np.sin(angulos)  # Seno de cada elemento
```

```python
arr = np.array([2, 3, 5])
np.sqrt(arr)  # Raiz quadrada de cada elemento
```

```python
np.exp(arr)   # Exponencial de cada elemento
```

### Manipulação de dimensões

```python
arr = np.array([1, 2, 3, 4, 5, 6])
arr = arr.reshape((2, 3))  # Array 2x3
arr
```

```python
flat_arr = arr.flatten() # converter para 1D
flat_arr
```

### Concatenar e dividir arrays

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
concatenado = np.concatenate((arr1, arr2))
concatenado
```

```python
arr = np.array([1, 2, 3, 4, 5, 6])
split_arrays = np.split(arr, 3)
split_arrays
```

### Condicionais com arrays

```python
arr = np.array([1, 2, 3, 4, 5])
arr[arr > 2]  # Elementos maiores que 2
```
