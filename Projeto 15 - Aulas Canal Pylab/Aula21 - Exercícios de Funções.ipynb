#### 4 - Reverso do número. Faça uma função que retorne o reverso de um número inteiro informado. Por exemplo: 127 -> 721.

```python
def reverso(num):
    return int(num[::-1])

x = input('Diga um número: ')

print(f'O reverso é {reverso(x)}')
```

#### 5 - Faça um programa em que uma função principal chama outras funções para calcular o valor máximo, o mínimo e a média dos elementos da lista a seguir:

    lista = [10, 20, 30, 40, 50]

```python
def minimo(x):
    return min(x)

def maximo(x):
    return max(x)

def media(x):
    return sum(x) / len(x)

def calcula_tudo(x):
    med = media(x)
    maior = maximo(x)
    menor = minimo(x)

    return med, maior, menor

lista = [1, 2, 3, 4, 5]

calcula_tudo(lista)
```

#### 8 - Avalie a lista de notas abaixo, classificando cada nota em "Aprovado" ou "Reprovado". Nota >= 7 significa aprovado. Cada nota deve estar acompanhada da sua classificação.

    lista = [5, 7.5, 9, 8, 7]

```python
def categoria(y):
    return 'Aprovado' if y >= 7 else 'Reprovado'

def avalia(x):
    tupla = [(i, categoria(i)) for i in x]
    return tupla

lista = [5, 7.5, 9, 8, 7]
avalia(lista)
```

#### 9 - Detector de Alienígenas: Crie um código que determina se o usuário é humano ou alienígena com base em dois critérios absurdos, que são se o nome dele tem mais de 10 letras e mais de 5 vogais.

```python
def alien_ou_humano(nome):
    conta_vogal = 0
    conta_letra = 0
    for i in nome:
      if i == 'a' or 'b' or 'c' or 'd' or 'u':
        conta_vogal +=1
      if i != ' ':
        conta_letra += 1

    return 'Alien' if conta_letra > 10 and conta_vogal > 5 else 'Humano'


alien_ou_humano('Tainan')
```

#### 10 - Sabemos que o primeiro ano de vida de um gato equivale a 15 anos humanos, o segundo a 10 anos, e os seguintes a 4 anos cada. Escreva uma função que receba a idade de um gato e retorne a idade correspondente em anos humanos.

```python
def idade_gato(idade):
    if idade == 1:
       return 'Equivale a 15 anos humanos.'
    elif idade == 2:
       return 'Equivale 25 anos humanos.'
    else:
       nova_id = 25 + (idade - 2) * 4
       return f'Equivale a {nova_id} anos humanos.'

idade = int(input('Diga a idade: '))
print(idade_gato(idade))
```
